```python
## Parte 1 — Resolução Manual

Sistema dado:

2x + y - z = 8
-3x - y + 2z = -11
-2x + y + 2z = -3
```

```python
### Matriz aumentada

[  2   1  -1 |  8 ]
[ -3  -1   2 | -11 ]
[ -2   1   2 | -3 ]
```

```python
### Etapa 1 — Zerando abaixo do pivô (2)

L2 = L2 + (3/2)L1
L3 = L3 + (2/2)L1  → L3 = L3 + L1
```

```python
Nova matriz:

[ 2   1  -1 |  8 ]
[ 0  0.5 0.5 | 1 ]
[ 0   2   1 | 5 ]
```

```python
### Etapa 2 — Zerando abaixo do segundo pivô

L3 = L3 - 4L2
```

```python
### Matriz triangular

[ 2   1  -1 |  8 ]
[ 0  0.5 0.5 | 1 ]
[ 0   0  -1 | 1 ]
```

```python
### Substituição regressiva

Da terceira equação:

- z = 1
z = -1

Substituindo na segunda:

0.5y + 0.5(-1) = 1
0.5y - 0.5 = 1
0.5y = 1.5
y = 3

Substituindo na primeira:

2x + 3 - (-1) = 8
2x + 4 = 8
2x = 4
x = 2
```

```python
### Solução do sistema

x = 2
y = 3
z = -1
```

```python
import numpy as np
```

```python
def gauss(A, b):
    n = len(b)

    # Monta a matriz aumentada
    M = np.hstack([A.astype(float), b.reshape(-1,1)])

    # Etapa 1: Eliminação
    for i in range(n):
        for j in range(i+1, n):
            fator = M[j][i] / M[i][i]
            M[j] = M[j] - fator * M[i]

    # Etapa 2: Substituição regressiva
    x = np.zeros(n)

    for i in range(n-1, -1, -1):
        soma = sum(M[i][j] * x[j] for j in range(i+1, n))
        x[i] = (M[i][-1] - soma) / M[i][i]

    return x
```

```python
A = np.array([
    [2, 1, -1],
    [-3, -1, 2],
    [-2, 1, 2]
])

b = np.array([8, -11, -3])

resultado = gauss(A, b)
print("Solução (Gauss):", resultado)
```

```python
resultado_numpy = np.linalg.solve(A, b)
print("Solução (NumPy):", resultado_numpy)
```

```python
A2 = np.array([
    [1, 1, 1],
    [2, 3, 1],
    [1, 2, 3]
])

b2 = np.array([6, 10, 13])

resultado2 = gauss(A2, b2)
print("Solução do desafio (Gauss):", resultado2)
```

```python
resultado2_numpy = np.linalg.solve(A2, b2)
print("Solução do desafio (NumPy):", resultado2_numpy)
```

## Parte 6 — Sistema com problema

Sistema:

x + y = 2  
2x + 2y = 4  

O que acontece:

O sistema possui equações proporcionais, ou seja, uma é múltiplo da outra.

Isso faz com que não exista solução única.

O algoritmo de Gauss apresenta problema (divisão por zero ou resultado indefinido).

Conclusão:

O sistema possui infinitas soluções (sistema possível indeterminado).

Como identificar:

Quando uma linha da matriz se torna toda zero durante a eliminação.

## Parte 4 — Análise

O resultado coincide com o obtido pelo numpy?

Sim, os resultados obtidos pelo método de Gauss coincidem com os resultados do numpy.linalg.solve, mostrando que a implementação está correta.

Qual etapa foi mais difícil de implementar?

A etapa mais difícil foi a eliminação, principalmente no controle dos índices e no cálculo dos fatores para zerar os elementos abaixo do pivô.

O método funciona para qualquer sistema?

Não. O método não funciona corretamente quando o sistema não possui solução única, como nos casos de sistemas com infinitas soluções ou sem solução.
