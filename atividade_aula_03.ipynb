```python
import numpy as np
import matplotlib.pyplot as plt
import math

def f(x):
    return math.cos(x) - x

a = 0
b = 1

x = np.linspace(a, b, 400)
y = [f(xi) for xi in x]

plt.figure(figsize=(8,5))
plt.plot(x, y, label='f(x) = cos(x) - x')
plt.axhline(0)
plt.grid(True)
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Gráfico da função cos(x) - x")
plt.show()
```

Existe mudança de sinal?
Sim.

O intervalo escolhido é adequado?
Sim, pois há mudança de sinal e isso garante que existe pelo menos uma raiz no intervalo.

```python
import pandas as pd

def bisseccao(f, a, b, n):

    dados = []

    for i in range(1, n+1):

        c = (a + b) / 2
        fc = f(c)

        dados.append([i, a, b, c, fc])

        if f(a) * fc < 0:
            b = c
        else:
            a = c

    tabela = pd.DataFrame(dados, columns=["Iteração","a","b","c","f(c)"])
    return tabela
```

```python
tabela = bisseccao(f, a, b, 10)
tabela
```

```python
raiz = tabela.iloc[-1]["c"]
print("Raiz aproximada:", raiz)
```

Quantas iterações foram necessárias para obter uma boa aproximação da raiz?

Após 10 iterações, a aproximação da raiz já é bastante precisa.

O intervalo diminuiu pela metade a cada iteração?

Sim. Esse é o princípio do método da bissecção: o intervalo é dividido em duas partes a cada passo.

O resultado obtido parece coerente com o gráfico da função?

Sim. O valor encontrado (~0.739) coincide com o ponto onde o gráfico cruza o eixo x.

Por que o método da bissecção sempre converge quando há mudança de sinal?

Porque se a função for contínua e existir mudança de sinal no intervalo, o Teorema do Valor Intermediário garante que existe pelo menos uma raiz dentro dele. O método reduz o intervalo pela metade a cada passo mantendo a raiz dentro dele.

Qual é a principal limitação desse método?

Ele pode ser mais lento que outros métodos, pois reduz o erro gradualmente.

Em quais situações ele pode ser mais adequado que outros métodos?

Quando se deseja um método simples, seguro e que sempre converge, mesmo sem uma boa aproximação inicial da raiz.
